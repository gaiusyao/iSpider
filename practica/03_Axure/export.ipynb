```python
import sqlite3
import pandas as pd
```

```python
sql = '''
    SELECT a.page_code
        , page_name
        , page_release
        , b.case_code
        , case_name
        , case_release
        , child_case_code
        , child_case_desc
    FROM page a
        , use_case b
        , child_case c
    WHERE b.page_code = c.page_code
        AND b.case_code = c.case_code
'''

df = pd.read_sql(sql, con=sqlite3.connect('axure.db'))
```
